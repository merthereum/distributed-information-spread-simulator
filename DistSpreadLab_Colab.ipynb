# DistSpreadLab — исследование распространения информации

Ноутбук запускает проект в Google Colab. Сначала загрузите ZIP-архив проекта, затем выполните остальные ячейки по порядку.

```python
from google.colab import files
import zipfile, os, pathlib

uploaded = files.upload()
zip_name = next(name for name in uploaded if name.lower().endswith('.zip'))
extract_dir = pathlib.Path('/content/distspreadlab_project')
extract_dir.mkdir(exist_ok=True)
with zipfile.ZipFile(zip_name, 'r') as zf:
    zf.extractall(extract_dir)

# Find the project root automatically.
project_candidates = list(extract_dir.rglob('run_project.py'))
if not project_candidates:
    raise FileNotFoundError('run_project.py не найден в архиве')
project_root = project_candidates[0].parent
os.chdir(project_root)
print('Project root:', project_root)
print('Files:', sorted(p.name for p in project_root.iterdir()))
```
